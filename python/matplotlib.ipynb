```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 50]

plt.plot(x, y, marker='o', linestyle='-', color='b', label="Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Line Plot")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))  # Set figure size
plt.plot(x, y, linestyle='--', marker='o', color='b', label='sine wave')
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Example Plot")
plt.legend()
plt.grid(True)
plt.savefig("example_plot.png", dpi=300)  # Save figure
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
x = np.linspace(0, 10, 100)  # Generates 100 points between 0 and 10
y = np.sin(x)  # Sine function

# Create the plot
plt.plot(x, y, color='blue', linestyle='--', linewidth=2, marker='o', markersize=5, label="Sine Wave")

# Customize the plot
plt.xlabel("X-axis (Time)")
plt.ylabel("Y-axis (Amplitude)")
plt.title("Basic Line Plot Example")
plt.legend()  # Show legend
plt.grid(True)  # Show grid
plt.show()  # Display the plot
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="Sine Wave", color='r', linestyle='-')
plt.plot(x, y2, label="Cosine Wave", color='b', linestyle='--')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plots Example")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='red', marker='o', alpha=0.6)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Scatter Plot")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 20, 15, 25, 30]

plt.bar(categories, values, color='blue')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Basic Bar Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)  # Generating random data

plt.hist(data, bins=30, color='purple', edgecolor='black', alpha=0.7)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Basic Histogram")
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 40, 10]
colors = ['blue', 'red', 'green', 'purple']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Basic Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = [np.random.randn(100) for _ in range(5)]  # 5 groups of random data

plt.boxplot(data, patch_artist=True)
plt.xlabel("Groups")
plt.ylabel("Values")
plt.title("Basic Box Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = [np.random.randn(100) for _ in range(5)]

plt.violinplot(data)
plt.xlabel("Groups")
plt.ylabel("Values")
plt.title("Basic Violin Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 10)
y = np.sin(x)
error = np.random.rand(10) * 0.2

plt.errorbar(x, y, yerr=error, fmt='o', color='red', ecolor='black', capsize=5)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Error Bar Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 1)
y = np.sin(x)

plt.stem(x, y, linefmt='r-', markerfmt='ro', basefmt='b-')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Stem Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.fill_between(x, y1, y2, color='gray', alpha=0.3)
plt.plot(x, y1, label="Sine", color='blue')
plt.plot(x, y2, label="Cosine", color='red')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Fill Between Lines")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 6)
y1 = [1, 4, 6, 8, 10]
y2 = [2, 2, 7, 10, 12]
y3 = [3, 5, 8, 12, 15]

plt.stackplot(x, y1, y2, y3, labels=['A', 'B', 'C'], colors=['r', 'g', 'b'])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Stack Plot")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000)
y = np.random.randn(1000)

plt.hexbin(x, y, gridsize=30, cmap='Blues')
plt.colorbar(label="Density")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Hexbin Plot")
plt.show()
```
